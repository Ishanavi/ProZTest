```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
import statistics
import random
import pandas as pd
import csv

df = pd.read_csv("medium_data.csv")
data = df["reading_time"].tolist()

fig = ff.create_distplot(
                            [data],
                            ["Reading Time"],
                            show_hist = False
)
fig.show()
```

```python
def random_set_of_mean(counter):
    dataset = []
    for i in range(0, counter):
        random_index= random.randint(0,len(data)-1)
        value = data[random_index]
        dataset.append(value)
    mean = statistics.mean(dataset)
    return mean
```

```python
def showFig(mean_list):
  fig = ff.create_distplot(
                            [mean_list],
                            ["Reading Time"],
                            show_hist = False
  )
  fig.show()
```

```python
mean_list = []
for i in range(0,100):
    set_of_mean = random_set_of_mean(30)
    mean_list.append(set_of_mean)
showFig(mean_list)



std_deviation = statistics.stdev(mean_list)
mean = statistics.mean(mean_list)
print("mean of sampling distribution:- ",mean)
print("Standard deviation of sampling distribution:- ", std_deviation)
```

```python
std_deviation_start_1, std_deviation_end_1 = mean-std_deviation, mean+std_deviation
std_deviation_start_2, std_deviation_end_2 = mean-(2*std_deviation), mean+(2*std_deviation)
std_deviation_start_3, std_deviation_end_3 = mean-(3*std_deviation), mean+(3*std_deviation)
```
